```python
import pandas as pd
df=pd.read_html('/content/DailyWiseMarketArrivals.html')[-1]
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
df.info()
```

```python
df.rename(columns={'Modal Price (Rs/q)':'y','Date':'ds'},inplace=True)
```

```python
df
```

```python
df_new=df[['ds','y']]
df_new
```

```python
df_new.info()
```

```python
df_new['ds']=pd.to_datetime(df_new['ds'])
df_new['y']=df_new['y'].astype(int)
```

```python
df_new
```

```python
df_new.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,5))
plt.scatter(df_new['ds'],df_new['y'])
```

```python
import prophet
op=prophet.Prophet()
op.fit(df_new)
```

```python
forecast =op.make_future_dataframe(periods=365,freq='D')
forecast =op.predict(forecast)
```

```python
op.plot(forecast,xlabel='Date',ylabel='Potato prices/q')
```
